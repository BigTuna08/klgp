```python
# Ave complexity is 123.7958984375
# After 980000 evals
# Evaluated 10000 programs in 0.438 secs (43.8498 micro secs per eval)
# On training, max fit = 0.98, min is = 0.55
# On cv, max fit = 0.95, min is = 0.375

compl = []
times = []
max_tr_fits = []
min_tr_fits = []
max_cv_fits = []
min_cv_fits = []


with open(log_file) as f:
    since_blank = 0
    for line in f.readlines():
        
        
        if len(line.strip()) < 2: continue

        line = line.split()
        
        if len(line) < 2: continue
#         print(line[1], type(line[1]), type("complexity"), line[1].rstrip().lstrip() == "complexity")
        
#         for i, char in enumerate(line[1]):
#             print(i, char)
            
        
#         for i, char in enumerate("complexity"):
#             print(i, char)        
        
        if line[0] == "Evaluated":
            times.append(float(line[7]))
        elif line[1] == "training,":
            min_tr_fits.append(float(line[10]))
            max_tr_fits.append(float(line[5]))
        elif line[1] == "cv,":
            min_cv_fits.append(float(line[10]))
            max_cv_fits.append(float(line[5]))
        elif line[1] == "complexity":
            compl.append(float(line[3]))

```
